```python
from transformers import pipeline, AutoTokenizer
import pandas as pd
from tqdm import tqdm
import torch
tqdm.pandas()
import re
import html
```

```python
summarizer = pipeline("text2text-generation", model="google/long-t5-tglobal-base")
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")

df = pd.read_csv("ai_image_reviews_1500.csv")

def summarize_text(text):
  tokenized = tokenizer.encode(text, truncation=False)
  if len(tokenized) > 512:
    prompt = "summarize: " + text
    try:
      return summarizer(prompt, max_new_tokens=512, min_new_tokens=30, no_repeat_ngram_size=3)[0]["generated_text"]
    except Exception as e:
      print(f"Error summarizing: {e}")
      return text
  else:
    return text

df["summarization"] = df["text"].astype(str).progress_apply(summarize_text)
df.to_csv("ai_image_reviews_1500_summarization.csv", index=False)
```

```python
summarizer = pipeline("text2text-generation", model="google-t5/t5-large")

tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")

def clean(text):
    text = html.unescape(text)  # Decode HTML entities
    text = re.sub(r'https?://\S+', '', text)  # Elimina URLs
    text = re.sub(r'@\w+', '', text)  # Elimina mencions
    text = re.sub(r'#\w+', '', text)  # Elimina hashtags
    text = re.sub(r'[^\w\s.,!?\'\"]+', '', text)  # Elimina símbols estranys, emojis, etc.
    text = re.sub(r'\s+', ' ', text)  # Substitueix múltiples espais per un de sol
    return text.strip()


df = pd.read_csv("ai_image_reviews_1500.csv")
def summarize_text(text):
  clean_text = clean(text)
  tokenized = tokenizer.encode(clean_text, truncation=False)
  if len(tokenized) > 512:
    try:
      return summarizer(clean_text, max_new_tokens=300, min_new_tokens=30, no_repeat_ngram_size=3)[0]["generated_text"]
    except Exception as e:
      print(f"Error summarizing: {e}")
      return clean_text
  else:
    return clean_text

df["summarization"] = df["text"].astype(str).progress_apply(summarize_text)
df.to_csv("ai_image_reviews_t5_large_clean_summarization.csv", index=False)
```
